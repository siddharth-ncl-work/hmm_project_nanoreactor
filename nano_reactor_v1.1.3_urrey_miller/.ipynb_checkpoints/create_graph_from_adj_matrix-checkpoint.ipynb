```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from labels import lables
%matplotlib inline

adj_mat=[]
def readAdjMat(file):
    global adj_mat
    adj_mat=[]
    c=0
    for line in file:
        if c==0:
            c+=1
            continue 
        adj_mat.append(map(lambda x:float(x),line.split(',')[2:]))
    adj_mat=np.matrix(adj_mat)

def plotGraph(G,pos):
    plt.figure(figsize=(15,8))
    nx.draw_networkx(G,pos,with_labels=False)
    nx.draw_networkx_labels(G,pos,lables)
    plt.tight_layout()

file=open('adjancency_matrix.csv','r')
readAdjMat(file)
print adj_mat.shape
G=nx.from_numpy_matrix(adj_mat)
pos=nx.spring_layout(G)
plotGraph(G,pos)
for i in G.neighbors(62):print i
```
